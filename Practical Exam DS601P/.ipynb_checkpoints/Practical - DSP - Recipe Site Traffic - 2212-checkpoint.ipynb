# Data Validation
This dataset has 947 rows, 8 columns. I have validated all features and have made changes to some features after validation below:
- 'recipe': 947 unique values, same as the description. No cleaning is needed.
- 'calories': numeric values with 52 missing values. Used median imputation as the distribution is right-skewed. 
- 'carbohydrate': numeric values with 52 missing values. Used median imputation as the distribution is right-skewed. 
- 'sugar': numeric values with 52 missing values. Used median imputation as the distribution is right-skewed. 
- 'protein': numeric values with 52 missing values. Used median imputation as the distribution is right-skewed. 
- 'category': 10 possible categories but dataset has 11 categories. Grouped 'Chicken Breast' with 'Chicken'. 
- 'servings': numeric values with some values as strings. Changed '4 as a snack' to '4' and '6 as a snack' to '6'. Changed dtype to int64.
- 'high_traffic': no missing values. Converted 'High' to 1 and NaN to 0 to prepare for modelling after.

# Model Development

Predicting the binary class of whether the recipe brings in high traffic or not is a classification problem in machine learning. I am choosing **Logistic Regression** because it a straightforward model that can provide a good benchmark for this classification problem even though it has been concluded that there are no or weak correlation among the numerical features. For the comparison model, I'll be using **Gradient Boosting Machines (GBM)** with decision tree as the base learner. 

For the evaluation metrics, I am choosing a single-number metric **F1 score**. The supporting metrics are **Accuracy** and **Precision**. The reasons for choosing these metrics are as follow:
- F1 Score: It takes into account both Precision and Recall. 
- Accuracy: In order to know the proportion of correctly classified values. It is a supporting metric as it can be **misleading with imbalanced data.**
- Precision: A lower proportion of false positives is better than a lower proportion of false negatives. If a low traffic recipe that is incorrectly identified as a high traffic recipe is being selected to be featured on the homepage, it would be worse than incorrectly predicting a high traffic recipe as a low traffic recipe.

## Prepare Data for Modelling 
To enable modelling, I chose to include all the features except 'recipe' as it is a unique identifier and does not contribute to the model. The target variable is 'high_traffic'. These are the following transformations done to the features:

1. Drop Feature - Drop the feature as it is a unique identifer and does not contribut to the model.
    - 'recipe'
3. Normalization - Log Transformation of the numerical features due to its right-skewed distribution.
    - 'calories', 'carbohydrate', 'sugar', 'protein'
 
4. Feature Scaling - Scaling of the normalized numerical features so that they are on the same scale with a mean of 0 and a standard deviation of 1.
    - 'calories', 'carbohydrate', 'sugar', 'protein'
5. One Hot Encoding - One Hot Encoding categorical features
    - 'category', 'servings' 

```python
# Drop 'recipe' as it is just the unique identifier and doesn't contribute anything to the model
df.drop(columns='recipe', inplace=True)
```

```python
# Determining outliers in the numerical features using the IQR
def calculate_iqr(column):
    # Sort the data 
    sorted_data = np.sort(column)

    # Calculate Q1 and Q3 
    Q1 = np.percentile(sorted_data, 25)
    Q3 = np.percentile(sorted_data, 75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Calculate lower and upper bounds 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers 
    outliers = [x for x in column if x < lower_bound or x > upper_bound]
    num_outliers = len(outliers) 
    
    print(f'IQR: {IQR:.3f}')
    print(f'Lower Bound: {lower_bound:.3f}')
    print(f'Upper Bound: {upper_bound:.3f}')
    print(f'Number of Outliers: {num_outliers}')
    print(f'Outliers: {outliers}')
```

```python
df.describe()
```

```python
print('Variable: Calories')
print(calculate_iqr(df.calories))
print('\n')
print('Variable: Carbohydrate')
print(calculate_iqr(df.carbohydrate))
print('\n')
print('Variable: Sugar')
print(calculate_iqr(df.sugar))
print('\n')
print('Variable: Protein')
print(calculate_iqr(df.protein))
```

There are 58, 66, 82, and 88 outliers for the numerical features - calories, carbohydrate, sugar, and protein respectively. Dropping rows with the outliers would not be a good choice as our dataset is small. We have 947 samples and removing 58 to 88 samples of outliers means removing 6% to 9% of our dataset. Instead, normalization techniques can be used to adjust the distribution of these numerical features to ensure that it follows a normal distribution. As all the distributions are right-skewed (as shown above in the histograms), the log transformation will be used. 

```python
# Normalization of numerical features 
numeric = ['calories', 'carbohydrate', 'sugar', 'protein']
for column in numeric:
    df['log_' + column] = np.log1p(df[column]) # df_log_transformed = df.apply(lambda x: np.log1p(x))  # log1p is used to handle zero values
    # Drop the original features 
    df.drop(columns=column, inplace=True)   
```

```python
# Create subplots
fig, axes = plt.subplots(2,2,figsize=(15,10))
log_numeric = ['log_calories', 'log_carbohydrate', 'log_sugar', 'log_protein']

# Flatten axes for easy iteration 
axes = axes.flatten()

# Plot histograms 
for i, variable in enumerate(log_numeric):
    sns.histplot(data=df, x=variable, bins=40, alpha=0.2, kde=True, ax=axes[i]).set(title='Distribution of ' + variable)
```

The distributions of the normalized numerical features follows more closely to a normal (Gaussian) distribution. Next, standard scaling will be done to ensure that the different features contribute equally to the model by bringing them to a similar scale with a mean close to 0 and a standard deviation of 1 for each feature.

```python
# Scale the transformed numerical features 
scaler = StandardScaler()

log_numeric = ['log_calories', 'log_carbohydrate', 'log_sugar', 'log_protein']
for column in log_numeric: 
    df['scaled_' + column] = scaler.fit_transform(df[[column]])
    # Drop the normalized features 
    df.drop(columns=column, inplace=True)
```

```python
# Create subplots
fig, axes = plt.subplots(2,2,figsize=(15,10))
scaled_log_numeric = ['scaled_log_calories', 'scaled_log_carbohydrate', 'scaled_log_sugar', 'scaled_log_protein']

# Flatten axes for easy iteration 
axes = axes.flatten()

# Plot histograms 
for i, variable in enumerate(scaled_log_numeric):
    sns.histplot(data=df, x=variable, bins=40, alpha=0.2, kde=True, ax=axes[i]).set(title='Distribution of ' + variable)
```

Next, the categorical features - 'category' and 'servings' will be one hot encoded. 

```python
# One hot encode the categorical features 
onehotencoder = OneHotEncoder(drop='first', sparse_output=False)

categorical = ['category', 'servings']
encoded_df = onehotencoder.fit_transform(df[categorical])

encoded_columns = onehotencoder.get_feature_names_out(categorical)
encoded_df = pd.DataFrame(encoded_df, columns=encoded_columns)

encoded_df = pd.concat([df.drop(columns=categorical), encoded_df], axis=1)
```

```python
# Split the dataset into both training and testing dataset 
X = encoded_df.drop(columns='high_traffic')
y = encoded_df.high_traffic

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Baseline Model - Logistic Regression Model

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred2 = lr.predict(X_test)
y_train_pred2 = lr.predict(X_train)
```

```python
# Evaluate the model 
# Accuracy Score 
accuracy_train = accuracy_score(y_train, y_train_pred2)
print(f'Train Dataset - Accuracy: {accuracy_train:.3f}')

# Precision Score
precision_train = precision_score(y_train, y_train_pred2)
print(f'Train Dataset - Precision: {precision_train:.3f}')

# F1 Score 
f1_score_train = f1_score(y_train, y_train_pred2)
print(f'Train  Dataset - F1 Score: {f1_score_train:.3f}')

print('------------------------------------------------')

# Accuracy Score 
accuracy_test = accuracy_score(y_test, y_pred2)
print(f'Test Dataset - Accuracy: {accuracy_test:.3f}')

# Precision Score
precision_test = precision_score(y_test, y_pred2)
print(f'Test Dataset - Precision: {precision_test:.3f}')

# F1 Score 
f1_score_test = f1_score(y_test, y_pred2)
print(f'Test Dataset - F1 Score: {f1_score_test:.3f}')

# Extra evaluation metrics for learning purposes 
print('------------------------------------------------')
print('FOR LEARNING PURPOSES:')
# Confusion Matrix
conf_matrix_test = confusion_matrix(y_test, y_pred2)
print('Confusion Matrix: ')
print(conf_matrix_test)

# Classification Report 
class_report_test = classification_report(y_test, y_pred2)
print('\nClassification Report: ')
print(class_report_test)

# AUC Score from ROC Curve 
auc_score_test = roc_auc_score(y_test, y_pred2)
print(f'AUC Score: {auc_score_test:.3f}')
```

**Interpretations:**
- Accuracy: The accuracy on both train and test datasets is quite close. This indicates the model is performing consistently acorss both datasets and is not overfitting or underfitting significantly. This metric shows that the model is correct 76% of the time. 
- Precision: The precision on both datasets is also close. This suggests that the model has a consistent ability to correctly classify positive instances across both datasets. The precision slightly improved on the test set, which is a positive sign that the model generalizes well to unseen data. 
- F1 Score: The F1 score is almost identical for both datasets, indicating that the model is balancing precision and recall well on both datasets.

#### Tuning of Hyperparameters with RandomizedSearchCV

```python
lr = LogisticRegression()
param_distributions = {
    'C': [0.01, 0.1, 1, 10, 100, 200, 300],
    'class_weight': [None, 'balanced'],
    'penalty': ['l1', 'l2', 'elasticnet', None],
    'solver': ['liblinear', 'saga', 'lbfgs'],
    'max_iter': [100, 200, 500, 1000, 1500, 2000]
}

random_search = RandomizedSearchCV(
    estimator=lr,
    param_distributions=param_distributions, 
    cv=5,
    verbose=2,
    scoring='accuracy',
    random_state=42
)

random_search.fit(X_train, y_train)
```

# Evaluate by Business Criteria

To recall, the business objectives are: 
1. Predict which recipes will lead to high traffic? (Baseline should be 40% as observed)
2. Correctly predict high traffic recipes 80% of the time. (Accuracy to be >= 80%)
3. Recommend the next steps whatever the results

**KPIs:**
1. Therefore, we would consider using the F1 score to consistently be above 0.8 as a KPI. After training the model, evaluate it on a validation set. Ensure that the F1 score on the validation set is above 0.8. This can be a requirement in the model evaluation pipeline. When deploying the model, the F1 score on new data should be continuously monitored as well. Set up alerts if accuracy drops below 80%. 

2. Another KPI would be to monitor the precision score as high precision ensures that when the model predicts a recipe as high-traffic, it is correct most of the time. This minimizes false positives, which is crucial if the cost of misclassifying a non-high-traffic recipe as high-traffic is high. In our case, high traffic means above the baseline of 40%.

3. Compare website traffic before and after implementing the model. The initial value for the percent of traffic should be at least 40%.

# Recommendations

1. Establish a regular retraining schedule, such as monthly or quarterly, to refresh the model with the latest data. Adjust hyperparameters and features as needed based on changes in user behavior or recipe trends.

2. Implement a Feedback Loop for Continuous Learning. Collect data on incorrectly predicted high-traffic recipes, and use this data to incrementally update the model, or through regular retraining. This ensures the model stays relevant and accurate as new trends emerge.
